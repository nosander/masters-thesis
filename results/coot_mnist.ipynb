# COOT on MNIST/USPS example

This notebook gives a simple example of the COOT problem between two well-known datasets: MNIST and USPS. 

We recall that for two datasets $\mathbf{X} \in \mathbb{R}^{n \times d}, \mathbf{X'} \in \mathbb{R}^{n' \times d'}$ the COOT problem aims at finding two optimal transport maps $\boldsymbol{\pi^{s}}$ and $\boldsymbol{\pi^{v}}$ between the samples and the features that solve:

\begin{equation}
\underset{\begin{smallmatrix}\boldsymbol{\pi^{s}} \in \Pi(\mathbf{w},\mathbf{w'}) \\ \boldsymbol{\pi^{v}} \in \Pi(\mathbf{v},\mathbf{v'}) \end{smallmatrix}} {\min} \sum_{ijkl} \left|X_{ik}-X_{jl}\right|^{p} \pi^{s}_{ij} \pi^{v}_{kl}
\end{equation}

where $\mathbf{w},\mathbf{w'}$ are histograms on the samples and $\mathbf{v},\mathbf{v'}$ are histogram on the features.

In our example the samples are images and the features are the pixels of the images. We will have $n,n'=300$ and $d=784, d'=256$

```python
def get_size(fig, dpi=100):
    with NamedTemporaryFile(suffix='.png') as f:
        fig.savefig(f.name, bbox_inches='tight', dpi=dpi)
        height, width, _channels = imread(f.name).shape
        return width / dpi, height / dpi

def set_size(fig, size, dpi=100, eps=1e-2, give_up=2, min_size_px=10):
    target_width, target_height = size
    set_width, set_height = target_width, target_height # reasonable starting point
    deltas = [] # how far we have
    while True:
        fig.set_size_inches([set_width, set_height])
        actual_width, actual_height = get_size(fig, dpi=dpi)
        set_width *= target_width / actual_width
        set_height *= target_height / actual_height
        deltas.append(abs(actual_width - target_width) + abs(actual_height - target_height))
        if deltas[-1] < eps:
            return True
        if len(deltas) > give_up and sorted(deltas[-give_up:]) == deltas[-give_up:]:
            return False
        if set_width * dpi < min_size_px or set_height * dpi < min_size_px:
            return False
```

## Load the MNIST/USPS dataset

```python
data=sp.io.loadmat('../git_repos/COOT/data/mnist.mat')

Xtot1=data['xapp'].astype(np.float32)

Ytot1=data['yapp']
d1=Xtot1.shape[1]
Ytot1[Ytot1==10]=0


data=sp.io.loadmat('../git_repos/COOT/data/usps.mat')
Xtot2=(data['xapp'].astype(np.float32)+1)/2

Ytot2=data['yapp']
Ytot2-=1
d2=Xtot2.shape[1]

np.random.seed(1976)
```

```python
# We take 300 samples per class
nbperclass=300
ntot=nbperclass*10
nbc=10
```

```python
def get_data(x,y,nbperclass):
    
    xr=np.zeros((0,x.shape[1]))
    yr=np.zeros((0))
    
    for i in range(np.max(y).astype(int)+1):
        
        xi=x[y.ravel()==i,:]
        idx=np.random.permutation(xi.shape[0])
        
        xr=np.concatenate((xr,xi[idx[:nbperclass],:]),0)
        yr=np.concatenate((yr,i*np.ones(nbperclass)))
        
    return xr,yr
```

```python
# Get Data
xs,ys=get_data(Xtot1,Ytot1,nbperclass)
#selmnist=xs.sum(0)>0

xt,ys=get_data(Xtot2,Ytot2,nbperclass)
```

```python
# Calculate negative data
L = 2**8
xs_n = (L - 1) - xs

# Calculate unform distributed data
temp = xs+1-1
scale = np.random.uniform(1,255,xs.shape[0])
xs_ud = np.divide(temp, scale[:,None])

# Normalize Data
xs = xs/255
xs_n = xs_n/255
xs_ud_1 = xs_ud/255
xs_ud_2 = 100*xs_ud/xs_ud.sum(axis = 1)[:,None]
```

# Define Helper Functions

```python
def solver(xs, xt, ys):
    # set the weights on the features
    vs=xs.sum(axis=0) 
    vs/=vs.sum()
    vt=xt.sum(axis=0)
    vt/=vt.sum()
    # COOT solver
    Ts,Tv,_,_=cot.cot_numpy(xs,xt,v1=vs,v2=vt,niter=100,log=True,algo='sinkhorn',reg=.5e-2, algo2='sinkhorn',reg2=.5e-2)
    return Ts
```

```python
def solver_gw(xs, xt, ys):
    ad_src = ad.AnnData(np.eye(xs.shape[0], xs.shape[0]))
    ad_src.obsm['X_src'] = xs
    ad_src.obs['cluster'] = ys

    ad_tgt = ad.AnnData(np.eye(xs.shape[0], xs.shape[0]))
    ad_tgt.obsm['X_tgt'] = xt
    ad_tgt.obs['cluster'] = ys

    adata = ad.concat([ad_src, ad_tgt], join = 'outer', label='batch', index_unique = '-')

    # Gromov Wasserstein solver
    gwp = GWProblem(adata)
    gwp = gwp.prepare(key="batch", x_attr={"attr": "obsm", "key": 'X_src'}, y_attr={"attr": "obsm", "key": 'X_tgt'})
    gwp = gwp.solve(alpha=0.5, epsilon=.5e-2)
    T = gwp[('0', '1')].solution.transport_matrix
    return np.array(T)
```

```python
def T_plot(T,ys, nbc = 10, ntot = 3000):
    C = np.zeros((nbc,nbc))
    for i in range(ntot):
        for j in range(ntot):
            if T[i,j]:
                C[int(ys[i]),int(ys[j])]+=T[i,j]
    f = np.sum(np.diag(C))
    return f, C
```

## "Normal" Dataset

```python
T_coot = solver(xs, xt, ys)
```

```python
f_coot, C_coot = T_plot(T_coot, ys)
print('Coot: Find the good class in {:.2f}% '.format(100*f_coot))
```

```python
T_gw = solver_gw(xs, xt, ys)
f_gw, C_gw = T_plot(T_gw, ys)
print('Coot: Find the good class in {:.2f}% '.format(100*f_gw))
```
